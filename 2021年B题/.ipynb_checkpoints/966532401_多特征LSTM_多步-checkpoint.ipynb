```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
sns.set()
from datetime import datetime
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
from sklearn import metrics
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Dropout
import matplotlib.gridspec as gridspec
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_excel('监测点A.xlsx')
df
```

```python
df.drop(columns=['地点','模型运行日期','预测时间'],axis=1,inplace = True)
df
```

```python
X_data = df.copy()
X_data.drop(columns= ['SO2实测浓度'],axis=1,inplace =True)
```

```python
Y_data = df[['SO2实测浓度']]  # 提取需求数据
```

```python
X = np.zeros((X_data.shape[0],
              72,
              X_data.shape[-1]))

Y = np.zeros((Y_data.shape[0],
             72))

print(X.shape, Y.shape)
```

```python
#序列构造
for sample_index in range(0, X_data.shape[0]):
    for month_index in range(0, 72):
        if sample_index+month_index >= X_data.shape[0]:  # 超界判断
            break
        X[sample_index, month_index, :] = X_data.iloc[sample_index+month_index, :]

    for j in range(0,72):
        if sample_index+j >= Y_data.shape[0]:  # 超界判断
            break
        Y[sample_index,j] = Y_data.iloc[sample_index+j, :]

print(X)
print(Y)

print(X.shape)
print(Y.shape)
```

```python
X_train = X[:-72]
Y_train = Y[:-72]

print(len(X_train))
print(len(Y_train))

X_val = X[-72:]
Y_val = Y[-72:]

print(len(X_val))
print(len(Y_val))

X_mean = X_train.mean(axis=0)
X_std = X_train.std(axis=0)
Y_mean = Y_train.mean(axis=0)
Y_std = Y_train.std(axis=0)

X_train_norm = (X_train - X_mean) / X_std
Y_train_norm = (Y_train - Y_mean) / Y_std
X_val_norm = (X_val - X_mean) / X_std
Y_val_norm = (Y_val - Y_mean) / Y_std
```

```python
model = Sequential()

model.add(LSTM(32,
               input_shape=(X_train_norm.shape[1], X_train_norm.shape[-1]),
               return_sequences=True))
model.add(Dropout(0.1))  # 防止过拟合
model.add(LSTM(32, return_sequences=True))
model.add(Dropout(0.1))
model.add(LSTM(32))
model.add(Dropout(0.1))

model.add(Dense(72))

model.compile(loss='mse',
              optimizer='rmsprop')

model.summary()
```

```python
history = model.fit(X_train_norm, Y_train_norm,
                    epochs=30,
                    batch_size=128, 
                    validation_data=(X_val_norm, Y_val_norm))
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.plot(range(len(loss)), loss, 'b-', label='训练集损失')
plt.plot(range(len(loss)), val_loss, 'r-', label='测试集损失')
plt.legend(loc='best')
plt.show();
```

```python
model_pred = model.predict(X_val_norm)
val_pred = model_pred * Y_std + Y_mean  # 别忘了，数据进行了标准化处理，因此预测值需要处理，再计算R方

val_pred_20 = val_pred[:,0]
val_pred_20

Y_val_20 = Y_val[:,0]
Y_val_20
```

```python
Y_val_20
```

```python
val_pred_20
```

```python
# 计算R2
R2 = metrics.r2_score(Y_val_20, val_pred_20) 
print('Test R2:%.3f' % R2)
# RMSE
rmse = sqrt(mean_squared_error(Y_val_20, val_pred_20))
print('Test RMSE:%.3f' % rmse)
# wmape
Y_val = np.float32(Y_val_20)
Y_val = Y_val.reshape(-1,1)
wmape = sum(abs(val_pred_20 - Y_val_20)) / sum(abs(Y_val_20))
print('Test wmape:%.3f' % wmape)

# 实际值与预测值对比图
plt.figure(figsize=(10, 6),dpi = 80)
plt.plot(val_pred_20, 'b-', label='prediction')
plt.plot(Y_val_20, 'r-', label='true')
plt.title('需求预测/全特征/多步')
plt.xlabel('月份')
plt.ylabel('需求数量')
plt.legend(loc='best')
plt.show();
```

```python
plt.figure(figsize=(10, 6),dpi = 80)
plt.plot(range(0,12),val_pred[:,0], label='1',c = 'y')
plt.plot(range(1,13),val_pred[:,1], label='2',c = 'g')
plt.plot(range(2,14),val_pred[:,2], label='3',c = 'b')
plt.plot(range(0,12),Y_val_20, 'r-', label='true')
plt.title('20年需求预测/全特征/多步')
plt.xlabel('月份')
plt.ylabel('需求数量')
plt.legend(loc='best')
plt.show();
```

# 多特征

```python
X_data = df[['数量', '平均气温(℃)']]# 提取特征数据
Y_data = df[['数量']]  # 提取需求数据

X = np.zeros((X_data.shape[0],
              12,
              X_data.shape[-1]))

Y = np.zeros((Y_data.shape[0],
             3))

print(X.shape, Y.shape)

#序列构造
for sample_index in range(0, X_data.shape[0]):
    for month_index in range(0, 12):
        if sample_index+month_index >= X_data.shape[0]:  # 超界判断
            break
        X[sample_index, month_index, :] = X_data.iloc[sample_index+month_index, :]

    for j in range(0,3):
        if sample_index+j >= Y_data.shape[0]:  # 超界判断
            break
        Y[sample_index,j] = Y_data.iloc[sample_index+j, :]

print(X)
print(Y)

print(X.shape)
print(Y.shape)
```

```python
X_train = X[:-24]
Y_train = Y[12:-12]

print(len(X_train))
print(len(Y_train))

X_val = X[-24:-12]
Y_val = Y[-12:]

print(len(X_val))
print(len(Y_val))

X_mean = X_train.mean(axis=0)
X_std = X_train.std(axis=0)
Y_mean = Y_train.mean(axis=0)
Y_std = Y_train.std(axis=0)

X_train_norm = (X_train - X_mean) / X_std
Y_train_norm = (Y_train - Y_mean) / Y_std
X_val_norm = (X_val - X_mean) / X_std
Y_val_norm = (Y_val - Y_mean) / Y_std
```

```python
model = Sequential()

model.add(LSTM(32,
               input_shape=(X_train_norm.shape[1], X_train_norm.shape[-1]),
               return_sequences=True))
model.add(Dropout(0.1))  # 防止过拟合
model.add(LSTM(32, return_sequences=True))
model.add(Dropout(0.1))
model.add(LSTM(32))
model.add(Dropout(0.1))

model.add(Dense(3))

model.compile(loss='mse',
              optimizer='rmsprop')

model.summary()
```

```python
history = model.fit(X_train_norm, Y_train_norm,
                    epochs=30,
                    batch_size=128, 
                    validation_data=(X_val_norm, Y_val_norm))
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.plot(range(len(loss)), loss, 'b-', label='训练集损失')
plt.plot(range(len(loss)), val_loss, 'r-', label='测试集损失')
plt.legend(loc='best')
plt.show();

model_pred = model.predict(X_val_norm)
val_pred = model_pred * Y_std + Y_mean  # 别忘了，数据进行了标准化处理，因此预测值需要处理，再计算R方

val_pred_20 = val_pred[:,0]
val_pred_20

Y_val_20 = Y_val[:,0]
Y_val_20

# 计算R2
R2 = metrics.r2_score(Y_val_20, val_pred_20) 
print('Test R2:%.3f' % R2)
# RMSE
rmse = sqrt(mean_squared_error(Y_val_20, val_pred_20))
print('Test RMSE:%.3f' % rmse)
# wmape
Y_val = np.float32(Y_val_20)
Y_val = Y_val.reshape(-1,1)
wmape = sum(abs(val_pred_20 - Y_val_20)) / sum(abs(Y_val_20))
print('Test wmape:%.3f' % wmape)

# 实际值与预测值对比图
plt.figure(figsize=(10, 6),dpi = 80)
plt.plot(val_pred_20, 'b-', label='prediction')
plt.plot(Y_val_20, 'r-', label='true')
plt.title('需求预测/全特征/多步')
plt.xlabel('月份')
plt.ylabel('需求数量')
plt.legend(loc='best')
plt.show();
```
